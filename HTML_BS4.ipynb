# Project -  scrapping random data from a website

```python
url = "https://books.toscrape.com/"
response = requests.get(url)
soup = BeautifulSoup(response.content, 'html.parser')
books = soup.find_all('article', class_='product_pod')
data = []
for book in books:
    title = book.h3.a['title']  # Get title
    price = book.find('p', class_='price_color').text  # Get price
    availability = book.find('p', class_='instock availability').text.strip() 

    data.append((title, price, availability))
# Print the data
for item in data:
    print(item)

import csv

with open('books_data.csv', mode='w', newline='') as file:
    writer = csv.writer(file)
    writer.writerow(['Title', 'Price', 'Availability'])  
    writer.writerows(data)  # Write data rows
```
